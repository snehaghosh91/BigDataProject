```python
import pandas as pd
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
```

```python
queens_latlong = pd.read_csv('data/latlong/queens_latlong.csv', header=None)
bronx_latlong = pd.read_csv('data/latlong/bronx_latlong.csv', header=None)
brooklyn_latlong = pd.read_csv('data/latlong/brooklyn_latlong.csv', header=None)
manhattan_latlong = pd.read_csv('data/latlong/manhattan_latlong.csv', header=None)
statenisland_latlong = pd.read_csv('data/latlong/statenisland_latlong.csv', header=None)
```

```python
left, bottom, right, top = -74.28, 40.49, -73.69, 40.93
m = Basemap(projection='cyl',llcrnrlat=bottom, urcrnrlat=top, 
              llcrnrlon=left, urcrnrlon=right, lat_ts=bottom,resolution='i')
m.drawcoastlines()
m.fillcontinents(color='orange', lake_color='aqua')
x1, y1 = m(queens_latlong[1].values,queens_latlong[0].values)
x2, y2 = m(bronx_latlong[1].values,bronx_latlong[0].values)
x3, y3 = m(brooklyn_latlong[1].values,brooklyn_latlong[0].values)
x4, y4 = m(manhattan_latlong[1].values,manhattan_latlong[0].values)
x5, y5 = m(statenisland_latlong[1].values,statenisland_latlong[0].values)
p1 = m.scatter(x1,y1,marker='.',color='b',latlon=True,
          zorder=10)
p2 = m.scatter(x2,y2,marker='.',color='m',latlon=True,
          zorder=10)
p3 = m.scatter(x3,y3,marker='.',color='g',latlon=True,
          zorder=10)
p4 = m.scatter(x4,y4,marker='.',color='r',latlon=True,
          zorder=10)
p5 = m.scatter(x5,y5,marker='.',color='c',latlon=True,
          zorder=10)
plt.legend((p1, p2, p3, p4, p5),
           ('Queens', 'Bronx', 'Brooklyn', 'Manhattan', 'Staten Island'),
           scatterpoints=1,
           loc='upper left',
           ncol=1,
           fontsize=10)
plt.title("Scatter Plot of Crimes in New York Boroughs")
plt.show()
```
